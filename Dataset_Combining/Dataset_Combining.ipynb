```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
headlines = pd.read_csv('News_articles_dataset.csv', parse_dates=['Date'], index_col="Date")
headlines.head()
```

```python
apple_data = pd.read_csv('HistoricalData_APPLE.csv', parse_dates=['Date'], index_col="Date")
apple_data.head()
```

```python
apple_data.sort_values('Date',inplace=True)
apple_data.head()
```

```python
apple_data.reset_index()
```

```python
df = pd.merge(headlines, apple_data, how='inner', on=['Date'])
df
```

```python
df = df.sort_index(ascending=True, axis=0)
df
```

```python
df.to_csv("News and Historical data append.csv")
```
