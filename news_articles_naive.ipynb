```python
import pandas as pd
import sklearn
import numpy as np
df1=pd.read_csv("News_articles_dataset.csv",parse_dates=["Date"])
df2=pd.read_csv("HistoricalData_1623748350213.csv",parse_dates=["Date"])
```

```python

df1.isnull().sum()
```

```python
df1.sort_index(inplace=True)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df2.sort_values('Date',inplace=True)
```

```python
df2.reset_index()
```

```python
a=pd.merge(df1,df2,how='inner',on=['Date'])

a.index=a['Date']
```

```python
a["Close/Last"]=a["Close/Last"].str.replace('$', '').astype(float)
a["Open"]=a["Open"].str.replace('$', '').astype(float)
a["High"]=a["High"].str.replace('$', '').astype(float)
a["Low"]=a["Low"].str.replace('$', '').astype(float)
```

```python
a.head()
```

```python
a.drop('Date',axis=1,inplace=True)
```

```python
a.head()
```

```python
a['close_price'] = a['Close/Last'].shift(1) 
```

```python
a['closed_price_diff']=''
a['closed_price_diff']=a['Close/Last']-a['close_price']
```

```python
a.head()
```

```python
a['profit/loss']=''
a['profit/loss'] = a['closed_price_diff'].apply(lambda x: 1 if x>0 else 0)
```

```python
a.head()
```

```python
a.drop(['Open','Low','High','Volume','Close/Last','close_price','closed_price_diff'],axis=1,inplace=True)
```

```python
a.head()
```

```python
import nltk
from nltk.corpus import stopwords
import string
def text_cleaning(a):
    remove_punctuation=[char for char in a if char not in string.punctuation]
    remove_punctuation=''.join(remove_punctuation)
    return [word for word in remove_punctuation.split() if word.lower() not in stopwords.words('english')]
```

```python
print(a.iloc[:,0].apply(text_cleaning))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
c_vectorizer=CountVectorizer(analyzer=text_cleaning).fit(a['Headlines'])

c_vectorizer.vocabulary_
```

```python
headlines_c=c_vectorizer.transform(a['Headlines'])
print(headlines_c)
```

```python
X=headlines_c.toarray()
print(X)
X.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
T_transformer=TfidfTransformer().fit(headlines_c)
print(T_transformer)
```

```python
Headlines_T=T_transformer.transform(headlines_c)
print(Headlines_T)
print(Headlines_T.shape)
```

```python
y=a['profit/loss']
```

```python
x_train = Headlines_T[:80]
y_train = y[:80]
x_test= Headlines_T[80:]
y_test= y[80:]
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
